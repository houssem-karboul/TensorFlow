### Manipulating Tensors (tensors operations)
If data is stored in tensors, finding patterns within these tensors often involves manipulating tensors. When building models with TensorFlow, much of the pattern discovery within tensors is done for you; however, oftentimes that pattern discovery is through the extended use of a few basic operations !

Addition operator:

```python
tensor=tf.constant([[10,7],
                    [3,4]])
tensor+10
tf.add(tensor,10) # With Tensorflow built-in function (executed with GPU !)
```

Multiplication operator(with value):


```python
tensor*10
tf.multiply(tensor,10) # With Tensorflow built-in function (executed with GPU !)
```

Substriction operator:

```python
tensor-1
tf.subtract(tensor,1) # With Tensorflow built-in function (executed with GPU !)
```

Divisional operator:


```python
tensor/10
tf.divide(tensor,10) # With Tensorflow built-in function (executed with GPU !)
```

Matrix multiplication (Also known as dot product)


```python
M1=tf.constant([[1,2,3],
    [4,5,6]]);
M2=tf.constant([[5,6],
 [7,8],
 [9,10]])
multiplication_result=tf.matmul(M1,M2); # multiplication using Tensorflow (GPU)
multiplication_result
```

### Aggregating tensors (condensing them from multiple values down to a smaller amount of val) :



```python
# Get the absolute values
simple_tensor=tf.constant([[-10,-7],[7,10]])
tf.abs(simple_tensor)
```

Let's go through the rollowing forms of aggregation:

	*Get the minimum
	*Get the maximum
	*Get the mean of a tensor
	*Get the sum of a tensor


```python
# Create random tensor:
simple_random_tensor=tf.constant(np.random.randint(-100,100,900))
simple_random_tensor.shape,simple_random_tensor.ndim, tf.size(simple_random_tensor)
tf.reshape(simple_random_tensor, shape=(6,6,25))
```

```python
# minimum
simple_tensor_min=tf.reduce_min(simple_random_tensor)
simple_tensor_min
```

```python
# maximum
simple_tensor_max=tf.reduce_max(simple_random_tensor)
simple_tensor_max
```

```python
# mean 
simple_tensor_mean =tf.reduce_mean(simple_random_tensor)
simple_tensor_mean 
```

```python
# sum  
simple_tensor_sum  =tf.reduce_sum(simple_random_tensor)
simple_tensor_sum 
```

```python
# variance
simple_random_tensor=tf.cast(simple_random_tensor,dtype=tf.float32) # convert the tensor to a floating-point type explicitly before calculating the variance to avoid numerical instability during the conversion and subsequent variance calculation.
tf.math.reduce_variance(simple_random_tensor)
```

```python
# standard deviation
tf.math.reduce_std(simple_random_tensor)
```

### One-hot encoding tensors
In order to transmit the data to our neural network, we need to encode it using numerical encoding. One-hot encoding, being a type of numerical encoding, serves as an effective method for this purpose.

```python
# Create a list of indices
my_list = [1, 2, 3, 4, 5]
numpy_array = np.array(my_list)
numpy_array
```

```python
# One hot encode our list of indices
tf.one_hot(numpy_array, depth=5)
```

```python
# One hot encode our list using offValue and onValue
tf.one_hot(numpy_array, depth=5,on_value=1, off_value=-1 )
```
